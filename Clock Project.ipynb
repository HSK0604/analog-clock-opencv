```python
import cv2
import numpy as np
import math
from datetime import datetime as dt

c = 250
p = math.pi
r1 = 200
r2 = 170
r3 = 190

window = np.zeros((500,500,3),dtype=np.uint8)
cv2.namedWindow('Clock')
cv2.circle(window,(250,250),200,(0,0,0),-1)

for i in range(60):
    angle = (i*p)/30
    x1 = int(c + r1 * np.sin(angle))
    y1 = int(c - r1 * np.cos(angle))
    x3 = int(c + r3 * np.sin(angle))
    y3 = int(c - r3 * np.cos(angle))
    
    if i%5==0:
        x2 = int(c + r2 * np.sin(angle))
        y2 = int(c - r2 * np.cos(angle))
        
        cv2.line(window,(x1,y1),(x2,y2),(0,255,255),thickness=4)
            
    else:
        cv2.line(window,(x1,y1),(x3,y3),(255,255,255),thickness=4)
        
    
for i in range(1,13):
    angle = (i*p)/6
    x = int(c + 150 * np.sin(  angle))
    y = int(c - 150 * np.cos( angle))
    cv2.putText(window,str(i),(x-10,y+10),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,0),2)
    
win = window.copy()
while True:
    window = win.copy()
    current_time = dt.now().strftime("%H:%M:%S")
    
    cv2.putText(window, current_time, (10, 40),cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
    
    # seconds hand
    sec=dt.now().second
    sin = int(c + 170 * np.sin((sec * p) / 30))
    cos = int(c - 170 * np.cos((sec * p) / 30))
    cv2.line(window,(250,250),(sin,cos),(0,255,0),thickness=2)
    
    # minutes hand
    minu=dt.now().minute
    sin = int(c + 140 * np.sin((minu * p) / 30))
    cos = int(c - 140 * np.cos((minu * p) / 30))
    cv2.line(window,(250,250),(sin,cos),(255,255,255),thickness=4)
    
    # hours hand
    hrs = dt.now().hour
    hrs = hrs % 12
    minu = dt.now().minute
    hrs = 5 * hrs + int(minu/12)
    sin = int(c + 80 * np.sin((hrs * p) / 30))
    cos = int(c - 80 * np.cos((hrs * p) / 30))
    cv2.line(window,(250,250),(sin,cos),(0,0,255),thickness=6)

    cv2.imshow('Clock',window)
    if cv2.waitKey(1)==ord('q'):
        break
cv2.destroyAllWindows()
```
